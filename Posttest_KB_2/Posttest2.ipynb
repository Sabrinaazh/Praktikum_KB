### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
jml_record = len(df)
jml_atribut = len(df.columns)
nm_atribut = list(df.columns)
jml_atribut_angka = len(df.select_dtypes(include=['int64', 'float64']).columns)

info_dict = {
    'Jumlah Record': jml_record,
    'Jumlah Atribut': jml_atribut,
    'Nama Atribut': nm_atribut,
    'Jumlah Atribut Dengan Tipe Data Angka': jml_atribut_angka
}

table = PrettyTable()
table.field_names = ['Informasi', 'Value']
for key, value in info_dict.items():
    if key == 'Nama Atribut':
        value = '\n'.join(value)
    table.add_row([key, value])
print(table)
```
